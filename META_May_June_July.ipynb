```python
import pandas as pd

# Replace with the actual filename if different
post_df = pd.read_csv("May-01-2025_Jul-28-2025.csv")
```

```python
# Convert 'Publish time' to datetime
post_df["Publish time"] = pd.to_datetime(post_df["Publish time"])

# Extract just the date
post_df["Date"] = post_df["Publish time"].dt.date

# Prepare your follows dataset (pasted or uploaded earlier)
from io import StringIO

follows_data = """
Date,Primary
2025-06-30T00:00:00,19
2025-07-01T00:00:00,15
2025-07-02T00:00:00,18
2025-07-03T00:00:00,16
2025-07-04T00:00:00,33
2025-07-05T00:00:00,27
2025-07-06T00:00:00,33
2025-07-07T00:00:00,38
2025-07-08T00:00:00,36
2025-07-09T00:00:00,26
2025-07-10T00:00:00,35
2025-07-11T00:00:00,27
2025-07-12T00:00:00,27
2025-07-13T00:00:00,27
2025-07-14T00:00:00,34
2025-07-15T00:00:00,20
2025-07-16T00:00:00,19
2025-07-17T00:00:00,34
2025-07-18T00:00:00,18
2025-07-19T00:00:00,18
2025-07-20T00:00:00,13
2025-07-21T00:00:00,19
2025-07-22T00:00:00,23
2025-07-23T00:00:00,26
2025-07-24T00:00:00,24
"""

df_follows = pd.read_csv(StringIO(follows_data), parse_dates=["Date"])
df_follows.rename(columns={"Primary": "Follows"}, inplace=True)
df_follows["Date"] = df_follows["Date"].dt.date

# Merge on the same posting date
merged_df = pd.merge(post_df, df_follows, on="Date", how="left")

# Preview merged results
merged_df[["Title", "Publish time", "Date", "Follows"]].head()
```

```python
top_follow_posts = merged_df.sort_values("Follows", ascending=False)[["Title", "Publish time", "Follows", "Post type", "Caption type"]]
top_follow_posts.head(10)
```

```python
# Create a copy to avoid modifying original DataFrame
top_df = top_follow_posts.copy()

# Clean the Title: remove hashtags, split to first 5 words
def clean_title(row):
    words = row["Title"].split()
    words = [w for w in words if not w.startswith("#")]
    short_title = " ".join(words[:5])
    return f"{row['Publish time'].date()} - {short_title}"

# Apply to dataframe
top_df["Cleaned Title"] = top_df.apply(clean_title, axis=1)

# Plot
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 8))
sns.barplot(data=top_df, x="Follows", y="Cleaned Title", palette="crest")
plt.title("Top Posts by Follower Gain (First 5 Words + Date)")
plt.xlabel("Follows")
plt.ylabel("Post (Date - First 5 Words)")
plt.gca().spines[['top', 'right']].set_visible(False)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Keep only numeric columns of interest
engagement_cols = ["Views", "Reach", "Reactions", "Comments", "Shares", "Total clicks", "Follows"]
corr_matrix = merged_df[engagement_cols].corr()

plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation between Engagement Metrics and Follows")
plt.show()
```

```python
# Create just a 'Date' column
top_df["Date"] = pd.to_datetime(top_df["Publish time"]).dt.date

plt.figure(figsize=(10, 5))
sns.lineplot(data=top_df, x="Date", y="Follows", marker="o")
plt.title("Follower Gain Over Time")
plt.xlabel("Post Date")
plt.ylabel("Follows")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=top_df, x="Post type", y="Follows", palette="Set2")
plt.title("Follower Gain by Post Type")
plt.xlabel("Post Type")
plt.ylabel("Follows")
plt.tight_layout()
plt.show()
```
